```python
import warnings
warnings.filterwarnings('ignore')
from google.colab import drive 
import pandas as pd
import numpy as np


drive.mount('/drive')
df=pd.read_csv('/drive/My Drive/Colab Notebooks/decision_theory/Indian Liver Patient Dataset (ILPD).csv',header = None)
df.columns =['age', 'gender', 'tb', 'db', 'tp', 'albumin', 'A/G ratio', 'SGPT', 'SGOT', 'Alkphos','healthy']
# view summary of dataset
df.info()
```

```python
df.shape
```

Filling of the nan rows with the mean of the column 

```python
#Finding the mean of the column having NaN
mean_value=df['Alkphos'].mean()
  
# Replace NaNs in column Alkphos with the
# mean of values in the same column
df['Alkphos'].fillna(value=mean_value, inplace=True)
df.info()
```

```python
#Convert Sex into numeric values (male -> 0 | Female -> 1)
dummy = pd.get_dummies(df['gender'])
df = pd.concat((dummy,df),axis =1 )
df = df.drop(['gender'],axis=1)
df = df.drop(['Female'],axis=1)
df = df.rename(columns={"Male": "gender"})
```
